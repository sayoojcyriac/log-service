```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data from the Excel file
df = pd.read_excel('/Users/sayoojcyriac/tech-career/jupyter/data-visualization/fids.xlsx')
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
for study in df['Study'].unique():
    study_data = df[df['Study'] == study]
    plt.plot(study_data.index, study_data['FID'], marker='o', label=study)

plt.title('FID (First Image Display) by Study')
plt.xlabel('Sample Index')
plt.ylabel('FID (ms)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar Plot
avg_fid = df.groupby('Study')['FID'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(avg_fid['Study'], avg_fid['FID'], color=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Average FID (First Image Display) by Study')
plt.xlabel('Study')
plt.ylabel('Average FID (ms)')
plt.show()
```

```python
# Simulated data for 10 studies with 10 samples each
data = {
    'Study': [f'Study-{i+1}' for i in range(10) for _ in range(10)],
    'Sample': [i+1 for _ in range(10) for i in range(10)],
    'FID (ms)': [1200 + i * 50 + j * 5 for i in range(10) for j in range(10)]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Line Plot with subplots
plt.figure(figsize=(15, 20))

for i, study in enumerate(df['Study'].unique(), start=1):
    study_data = df[df['Study'] == study]
    plt.subplot(5, 2, i)
    plt.plot(study_data['Sample'], study_data['FID (ms)'], marker='o', label=study)
    plt.title(f'FID Times for {study}')
    plt.xlabel('Sample Index')
    plt.ylabel('FID (ms)')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Bar Plot with color coding
avg_fid = df.groupby('Study')['FID (ms)'].mean().reset_index()

plt.figure(figsize=(15, 8))
colors = plt.cm.tab20.colors  # Use a colormap for distinct colors
for i, study in enumerate(avg_fid['Study']):
    plt.bar(study, avg_fid.loc[i, 'FID (ms)'], color=colors[i % 20])

plt.title('Average FID (First Image Display) Times by Study')
plt.xlabel('Study')
plt.ylabel('Average FID (ms)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Combined Line Plot
plt.figure(figsize=(15, 10))
markers = ['o', 'v', 's', 'P', '*', 'X', 'D', '^', '<', '>']
colors = plt.cm.tab10.colors  # Use a colormap for distinct colors

for i, study in enumerate(df['Study'].unique()):
    study_data = df[df['Study'] == study]
    plt.plot(study_data['Sample'], study_data['FID (ms)'], marker=markers[i % len(markers)], color=colors[i % len(colors)], label=study)

plt.title('FID (First Image Display) Times by Study')
plt.xlabel('Sample Index')
plt.ylabel('FID (ms)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
# Combined Bar Plot with grouped bars
avg_fid = df.groupby(['Study', 'Sample'])['FID (ms)'].mean().unstack()

plt.figure(figsize=(15, 8))
colors = plt.cm.tab10.colors  # Use a colormap for distinct colors
bar_width = 0.08
index = np.arange(avg_fid.shape[1])

for i, study in enumerate(avg_fid.index):
    plt.bar(index + i * bar_width, avg_fid.loc[study], bar_width, label=study, color=colors[i % len(colors)])

plt.title('Average FID (First Image Display) Times by Study')
plt.xlabel('Sample Index')
plt.ylabel('Average FID (ms)')
plt.xticks(index + bar_width * (len(avg_fid.index) - 1) / 2, avg_fid.columns)
plt.legend()
plt.show()
```

```python
# Combined Bar Plot with grouped bars
avg_fid = df.groupby(['Study', 'Sample'])['FID (ms)'].mean().unstack()

plt.figure(figsize=(15, 8))
colors = plt.cm.tab10.colors  # Use a colormap for distinct colors
bar_width = 0.08
index = np.arange(avg_fid.shape[1])

for i, study in enumerate(avg_fid.index):
    plt.bar(index + i * bar_width, avg_fid.loc[study], bar_width, label=study, color=colors[i % len(colors)])

plt.title('Average FID (First Image Display) Times by Study')
plt.xlabel('Sample Index')
plt.ylabel('Average FID (ms)')
plt.xticks(index + bar_width * (len(avg_fid.index) - 1) / 2, avg_fid.columns, rotation=45, ha='right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
